```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Users\narla\Downloads\Healthcare project\hospital_readmissions.csv")

df.head()
```

```python
df.columns = df.columns.str.lower().str.strip()
```

```python
df['readmitted_flag'] = df['readmitted'].map({'yes': 1, 'no': 0})
```

```python
def convert_age(age):
    age = age.strip('[]()')
    lower = int(age.split('-')[0])
    upper = int(age.split('-')[1])
    return (lower + upper) / 2

df['age_numeric'] = df['age'].apply(convert_age)
```

```python
df['medical_specialty'] = df['medical_specialty'].replace("Missing", np.nan)
```

```python
df['total_visits'] = df['n_outpatient'] + df['n_inpatient'] + df['n_emergency']
```

```python
df['high_utilization_flag'] = (
    df['total_visits'] > df['total_visits'].quantile(0.75)
).astype(int)
```
